# AirBnB Dataset

```python
import pandas as pd
```

```python
data = pd.read_csv("Airbnb_Open_Data.csv")
```

```python
data.head()
```

```python
data.set_index("id")
```

```python
data.dtypes.value_counts()
```

```python
data.info()
```

```python
data = data.drop(columns = ["NAME","host name","country","country code","last review","license","house_rules","number of reviews",
                            "last review","reviews per month","calculated host listings count","availability 365"])
```

```python
data = data.set_index("id")
```

```python
data.info()
```

```python
data.groupby(["neighbourhood group"])["neighbourhood group"].count()
```

```python
data["neighbourhood group"].replace({"brookln":"Brooklyn","manhatan":"Manhattan"},inplace=True)
```

```python
data.groupby(["neighbourhood group"])["neighbourhood group"].count()
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
data 
```

```python
price = []
for i in data["price"]:
    i = i[1:]
    if len(i)>5:
        i = i[:-5]+i[-4:]
    price.append(i)
pd.to_numeric(price, downcast="integer")

ser_fee = []
for i in data["service fee"]:
    i = i[1:]
    if len(i)>5:
        i = i[:-5]+i[-4:]
    ser_fee.append(i)
pd.to_numeric(ser_fee, downcast ="integer")
```

```python
data["price"] = price
data["service fee"] = ser_fee

data["service fee"] = data["service fee"].astype(int)
data["price"] = data["price"].astype(int)
```

```python
data
```

```python
data = data[data["minimum nights"]>0]
data = data[data["minimum nights"]<365]
data.info()
```

```python
data.to_csv("Tertemiz.csv")
```
